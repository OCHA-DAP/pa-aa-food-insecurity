```python
def return_max_csa(admin1,admin2, df,status,adm2_col="ADM2_EN",adm1_col="ADM1_EN",adm0_col="ADM0_EN"):
    '''
    Return the maximum IPC value in an admin2 district.
    '''
    sub = df.loc[(df[adm1_col]==admin1)&(df[adm2_col]==admin2)]
    mx = sub['area'].max()
    row = sub[[adm0_col,adm1_col,adm2_col,status]].loc[sub['area']==mx]
    return row
```

```python
def return_max_cs(admin1,admin2, df,status,adm2_col="ADM2_EN",adm1_col="ADM1_EN",adm0_col="ADM0_EN"):
    '''
    Return the maximum IPC value in an admin2 district.
    '''
    sub = df.loc[(df[adm1_col]==admin1)&(df[adm2_col]==admin2)]
    mx = sub['CS'].max()
    
    if np.isnan(mx):
        print(admin1)
        print(sub[status])
        print(sub[[adm0_col,adm1_col,adm2_col,status]])#.loc[sub['CS']==mx])
    row = sub[[adm0_col,adm1_col,adm2_col,status]].loc[sub['CS']==mx].iloc[0]
    return row
```

```python
def return_max_arcs(admin1,admin2, df,status,adm2_col="ADM2_EN",adm1_col="ADM1_EN",adm0_col="ADM0_EN"):
    '''
    Return the maximum IPC value in an admin2 district.
    '''
    sub = df.loc[(df[adm1_col]==admin1)&(df[adm2_col]==admin2)]
    if len(sub[sub[status]!=0])>0:
        sub=sub[sub[status]!=0]
    mx = sub['CS'].max()
    
    if np.isnan(mx):
        print(admin1)
        print(sub[status])
        print(sub[[adm0_col,adm1_col,adm2_col,status]])#.loc[sub['CS']==mx])
    row = sub[[adm0_col,adm1_col,adm2_col,status]].loc[sub['CS']==mx].iloc[0]
    return row
```

```python
def return_max_arc(admin1,admin2, df,status,adm2_col="ADM2_EN",adm1_col="ADM1_EN",adm0_col="ADM0_EN"):
    '''
    Return the maximum IPC value in an admin2 district.
    '''
    sub = df.loc[(df[adm1_col]==admin1)&(df[adm2_col]==admin2)]
    if len(sub[sub[status]!=0])>0:
        sub=sub[sub[status]!=0]
    mx = sub['CS'].max()
    
    if np.isnan(mx):
        print(admin1)
        print(sub[status])
        print(sub[[adm0_col,adm1_col,adm2_col,status]])#.loc[sub['CS']==mx])
    row = sub[[adm0_col,adm1_col,adm2_col,status]].loc[sub['CS']==mx].iloc[0]
    return row
```
