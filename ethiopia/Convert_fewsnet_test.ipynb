```python
status="ML2"
dffn = gpd.read_file("Data/EA_FewsNet/EA_201910_{}.shp".format(status))
```

```python
dffn.head()
```

```python
overlap = gpd.overlay(dfad, dffn, how='intersection')
```

```python
overlap = overlap.drop_duplicates()
overlap['area'] = overlap['geometry'].area
columns = ['ADMIN0', 'ADMIN1', 'ADMIN2','geometry', 'area',status]
overlap = overlap[columns]
```

```python
def return_max_cs(admin2, df,status):
    '''
    Return the maximum IPC value in an admin2 district.
    '''
    sub = df.loc[(df['ADMIN2']==admin2)]
    mx = sub['area'].max()
    row = sub[[ 'ADMIN0', 'ADMIN1', 'ADMIN2',status]].loc[sub['area']==mx]
    return row
```

```python
new_df = pd.DataFrame(columns=[ 'ADMIN0', 'ADMIN1', 'ADMIN2', status])
for a in overlap['ADMIN2'].unique():
    row = return_max_cs(a, overlap,status)
    new_df = new_df.append(row)
#     new_df.to_csv(COUNTRY + '_admin2_fewsnet_' + STATUS + str(today) + '.csv')
```
