```python
ad_columns=["ADM0_EN","ADM1_EN","ADM2_EN"]
```

```python
overlap = overlap.drop_duplicates()
overlap['area'] = overlap['geometry'].area
columns = ad_columns+['geometry', 'area',status]
overlap = overlap[columns]
```

```python
def return_max_cs(admin1,admin2, df,status,adm2_col="ADM2_EN",adm1_col="ADM1_EN",adm0_col="ADM0_EN"):
    '''
    Return the maximum IPC value in an admin2 district.
    '''
    sub = df.loc[(df[adm1_col]==admin1)&(df[adm2_col]==admin2)]
    mx = sub['area'].max()
    row = sub[[adm0_col,adm1_col,adm2_col,status]].loc[sub['area']==mx]
    return row
```

```python
new_df = pd.DataFrame(columns=ad_columns+[status])
for index,ad_com in overlap[['ADM1_EN',"ADM2_EN"]].drop_duplicates().iterrows():
    row = return_max_cs(ad_com["ADM1_EN"],ad_com["ADM2_EN"], overlap,status)
    new_df = new_df.append(row)
#     new_df.to_csv(COUNTRY + '_admin2_fewsnet_' + STATUS + str(today) + '.csv')
```
